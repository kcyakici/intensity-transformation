```python
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2gray
from numpy import copy, array


def applyIntensityTransformation(grayscaled_image, gamma):
    for i in range(len(grayscaled_image)):
        for j in range(len(grayscaled_image[i])):
            # original formula: s = c * r ** gamma, c is taken as 1 here
            grayscaled_image[i][j] = grayscaled_image[i][j] ** gamma


def plot_image(image_1, image_2, title_1="Original", title_2="New Image"):
    plt.figure(figsize=(10, 10))
    plt.subplot(1, 2, 1)
    io.imshow(image_1)
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    io.imshow(image_2)
    plt.title(title_2)
    plt.show()


def main():
    brood_war_overmind_image_url = "https://static.wikia.nocookie.net/starcraft/images/f/f6/Overmind_SC2_Head1.jpg/revision/latest/scale-to-width-down/328?cb=20151118232210"

    overmind_image_original = io.imread(brood_war_overmind_image_url)
    gray_overmind = rgb2gray(overmind_image_original)

    old_image = array(gray_overmind).copy()

    # here I choose 0.6 for gamma and pass it as argument
    applyIntensityTransformation(gray_overmind, 0.6)

    plot_image(old_image, gray_overmind)


if __name__ == "__main__":
    main()
```
